```python
import seaborn as sns
import matplotlib.pyplot as plt
```

Inflação: A década de 1980 foi marcada por altos níveis de inflação no Brasil, que atingiram taxas extremamente altas.

Crises econômicas: Além da inflação, o Brasil enfrentou crises econômicas durante esse período, incluindo a crise da dívida externa.

Clima: O clima desempenha um papel crucial na produção de vinho. Condições climáticas adversas, como secas ou chuvas em excesso, podem afetar a qualidade e a quantidade da colheita.

Mercado Internacional: O mercado internacional de vinho é altamente competitivo, e a qualidade e o preço desempenham um papel importante na capacidade de um país exportar vinho.

```python
# Filtro por tipo de vinho (Vinho De Mesa)
tipo_de_vinho = 'Vinho De Mesa'
df_filtrado = df[df['Tipo Vinho'] == tipo_de_vinho]

# Crie um gráfico usando Seaborn catplot
plt.figure(figsize=(10, 6))
sns.set(style="whitegrid")
g = sns.catplot(data=df_filtrado, x='Ano', y='Quantidade (L)', kind='bar', height=6, aspect=2)
g.fig.suptitle(f'Quantidade {tipo_de_vinho} Exportados por Ano', y=1.02)
plt.xlabel('Ano')
plt.ylabel('Quantidade (L)')

# Rotacionar os rótulos dos anos para torná-los mais legíveis
g.set_xticklabels(rotation=45)

plt.show()
```
