# Análises

```python
import pandas as pd
```

```python
dados = pd.read_csv("https://raw.githubusercontent.com/icarocarmona/tech_challenge_f1/main/dados/trusted/dados_de_vinhos.csv", sep=";" )
dados
```

```python
top_10_paises = dados.groupby(by='Destino')['Quantidade (L)'].sum().sort_values(ascending=False).head(10).index
top_10_paises
```

```python
destinosTop10 = dados[dados['Destino'].isin(top_10_paises)]
destinosTop10
```

```python
destinosTop10_Ano = destinosTop10[destinosTop10['Ano'] >= 2007]
destinosTop10_Ano
```

```python
import seaborn as sns
```

```python
sns.catplot(data=destinosTop10_Ano, x="Ano", y="Quantidade (L)", col="Destino", kind="bar")
```

```python
# Crie um gráfico usando Seaborn catplot
g = sns.catplot(data=destinosTop10_Ano, x="Ano", y="Quantidade (L)", col="Destino", kind="bar")

# Adicione a rotação dos rótulos dos anos em cada subgráfico
for ax in g.axes.flat:
    ax.set_xticklabels(ax.get_xticklabels(), rotation=45)

```
